### 6. Downloads data from a given URL in chunks converts it to CSV format

```python
def download_data_to_desktop(url, file_name, chunk_size, num_records=999999):
    """
    Downloads data from a given URL in chunks, converts it to CSV format,
    and saves each chunk as a CSV file on the desktop.

    Args:
        url (str): The URL to download the data from.
        file_name (str): The base name of the CSV file to be created (without the chunk number).
        chunk_size (int): The number of records to download in each chunk.
        num_records (int): The total number of records to download (default is 2 million).
    """
    
    # Calculate the number of chunks needed
    num_chunks = num_records // chunk_size + (0 if num_records % chunk_size == 0 else 1)

    # Loop through the number of chunks
    for i in range(num_chunks):
        offset = i * chunk_size
        print(f"Processing chunk {i+1} of {num_chunks}, records {offset} to {min(offset + chunk_size, num_records)}")

        try:
            # Getting the data for this chunk
            response = requests.get(url + f"?$limit={chunk_size}&$offset={offset}")
            response.raise_for_status()  # Raise an exception if the request was unsuccessful
            data = response.json()

            # Converting the data to a pandas DataFrame
            df = pd.DataFrame(data)

            # Define the path for the desktop, adjusting 'YourUsername' as needed
            desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop')

            # Define the full file path
            file_path = os.path.join(desktop_path, f'{file_name}_{i+1}.csv')

            # Converting the DataFrame to CSV and saving to desktop
            df.to_csv(file_path, index=False)
            print(f"Saved chunk {i+1} to {file_path}")

        except requests.exceptions.RequestException as e:
            print(f"Request to API failed: {e}")
            break  # If a request fails, exit the loop

    print("Download Complete")

```

### Determine the category of each column in a DataFrame.

```python
def get_var_category(series):
    """
    Determine the category of a given series (column) in a DataFrame.
    
    Parameters:
        series (pandas.Series): The series to determine the category of.
        
    Returns:
        str: The category of the series.
    """
    unique_count = series.nunique(dropna=False)
    total_count = len(series)
    
    # Check if the series is numeric
    if pd.api.types.is_numeric_dtype(series):
        return 'Numerical'
    
    # Check if the series is a datetime type
    elif pd.api.types.is_datetime64_dtype(series):
        return 'Date'
    
    # Check if all values in the series are unique
    elif unique_count == total_count:
        return 'Text (Unique)'
    
    # If none of the above conditions are met, consider the series as categorical
    else:
        return 'Categorical'

def print_categories(df):
    """
    Print the category of each column in a DataFrame.
    
    Parameters:
        df (pandas.DataFrame): The DataFrame to print the categories of.
    """
    for column_name in df.columns:
        # Print the column name and its corresponding category
        print(column_name, ": ", get_var_category(df[column_name]))
```

### Missing values proportion 


```python
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame and you want to replace NaN, NA, NAN, or blank with pd.NA
# Standardize missing values to np.nan (which is the standard missing value marker used by pandas)
df.replace([np.nan, pd.NA, 'NA', 'NAN', ''], np.nan, inplace=True)

# Initialize a dictionary to store the proportion of missing values for different missing indicators
missing_proportions = {'NaN': {}, 'NA': {}, 'NAN': {}, 'blank': {}}

# Loop through all columns in df
for column in df.columns:
    # Count the total values for proportion calculation
    total_values = df.shape[0]

    # Check for standard NaN values (which includes all the types after replacement)
    missing_values_count = df[column].isna().sum()
    proportion_missing = missing_values_count / total_values
    missing_proportions['NaN'][column] = proportion_missing

    # Print column-wise missing value statistics if proportion is greater than 0
    if proportion_missing > 0:
        print(f"Column '{column}':")
        print(f"  Count of missing values: {missing_values_count}")
        print(f"  Proportion of missing values: {proportion_missing:.2%}")
```

### Check the unique values in DataFrame df according to the Data dictionry variable information


```python
import pandas as pd


# Define the expected unique values for categorical columns based on your data dictionary
categorical_expected_values = {
    'vendorid': [1, 2],
    'ratecodeid': [1, 2, 3, 4, 5, 6, 99],
    'store_and_fwd_flag': ['Y', 'N'],
    'payment_type': [0, 1, 2, 3, 4, 5, 6],
}

# Function to verify and print unique values in categorical columns
def verify_categorical_columns(df, expected_values):
    for col, expected in expected_values.items():
        # Ensure the column exists in DataFrame
        if col not in df.columns:
            print(f"Column: {col} does not exist in the DataFrame.")
            continue

        unique_values = df[col].dropna().unique().tolist()
        unexpected_values = [v for v in unique_values if v not in expected]
        
        print(f"Column: {col}")
        print(f"Expected: {expected}")
        print(f"Actual: {unique_values}")
        
        if unexpected_values:
            print(f"Unexpected values: {unexpected_values}")
        else:
            print("All actual values are expected.")
        print()  # Print a blank line for readability

# Run the verification function on the DataFrame 'df'
verify_categorical_columns(df, categorical_expected_values)
```
